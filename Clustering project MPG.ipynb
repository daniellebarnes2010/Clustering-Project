```python
#Import Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
#Load Data
Mpg = sns.load_dataset('mpg')
Mpg.head()
```

```python
Mpg.info()

#drop string columns
MpgTrimmed = Mpg.drop(['origin','name'], axis=1)
```

```python
MpgTrimmed.head()
```

```python
MpgTrimmed.fillna(0, inplace = True)
```

```python
#Find kmeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(MpgTrimmed)
```

```python
plt.figure(figsize=(10,6))
plt.title('K Means')
plt.scatter(MpgTrimmed['horsepower'], MpgTrimmed['displacement'], c=kmeans.labels_, cmap='viridis')
```

```python
plt.figure(figsize=(10,6))
plt.title('K Means')
plt.scatter(MpgTrimmed['weight'], MpgTrimmed['horsepower'], c=kmeans.labels_, cmap='viridis')
```

```python
plt.figure(figsize=(10,6))
plt.title('K Means')
plt.scatter(MpgTrimmed['weight'], MpgTrimmed['acceleration'], c=kmeans.labels_, cmap='viridis')
```
